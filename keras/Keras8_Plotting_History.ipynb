### Training History

Using Keras Callbacks to display training history - get the training metrics for each epoch - loss and accuracy values for the various sets

```python
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
seed = 7
np.random.seed(seed)
 
dataset = pd.read_csv("pima-indians-diabetes.csv", header=None, prefix='V')

dataset.head()
```

```python
#seperate data into input and output variables
X = np.array(dataset.iloc[:, 0:8])
y = np.array(dataset.iloc[:, 8])
```

```python
#define a model
model  = Sequential()
model.add(Dense(12,input_dim=8,kernel_initializer='uniform',activation='relu'))
model.add(Dense(8,kernel_initializer='uniform',activation='relu'))
model.add(Dense(1,kernel_initializer='uniform',activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

history = model.fit(X,y,validation_split=0.33,epochs=150,batch_size=10,verbose=0)
```

Using the history callback we can obtain values for loss and accuracy and plot

```python
print(history.history.keys())
# summarize history for accuracy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'], loc='upper left')
plt.show()
```

Validation set is labeled as test - accuracy is still rising suggests can still train a little more

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train' ,  'test'], loc= 'upper left' )
plt.show()
```

```python
From loss plot, can see that both sets have comparable performance 
```
